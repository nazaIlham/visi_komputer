```python
image1 = cv2.imread('test.jpg') 
```

```python
img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
  
# applying different thresholding 
# techniques on the input image
# all pixels value above 120 will 
# be set to 255
ret, thresh1 = cv2.threshold(img, 50, 260, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 50, 260, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 50, 260, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 50, 260, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 50, 260, cv2.THRESH_TOZERO_INV)
  
# the window showing output images
# with the corresponding thresholding 
# techniques applied to the input images
cv2_imshow(image1)
cv2_imshow(thresh1)
cv2_imshow(thresh2)
cv2_imshow(thresh3)
cv2_imshow(thresh4)
cv2_imshow(thresh5)
    
# De-allocate any associated memory usage  
if cv2.waitKey(0) & 0xff == 27: 
    cv2.destroyAllWindows() 
```
